```python
#apple stocks
q1 = df1[df1.ticker == 'APPL'].set_index("date")
q1
```

```python
#closing price
q2 = df1[df1.ticker == 'AMZN']
q3 = df1[df1.ticker == 'GOOG']
q4 = df1[df1.ticker == 'MSFT']
```

```python
plt.rcParams['figure.figsize']=(15.0,8.0)
style.use('ggplot')
plt.subplots_adjust(hspace= 0.25)
plt.subplot(2,2,1)
plt.title('Apple Closing Price Analysis')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.plot(q1.date,q1.close,color= 'red')
plt.legend(q1.ticker)
plt.subplot(2,2,2)
plt.title('Amazon Closing Price Analysis')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.plot(q2.date,q2.close,color= 'b')
plt.legend(q2.ticker)
plt.subplot(2,2,3)
plt.title('Google Closing Price Analysis')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.plot(q3.date,q3.close,color= 'g')
plt.legend(q3.ticker)
plt.subplot(2,2,4)
plt.title('Microsoft Closing Price Analysis')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.plot(q4.date,q4.close,color= 'c')
plt.legend(q4.ticker)
plt.show()
```

```python
plt.rcParams['figure.figsize']=(15.0,8.0) 
style.use('ggplot')
plt.subplots_adjust(top=9,bottom=.123)
plt.subplot(4,1,1)
plt.title ('Apple Stock Volume Analysis')
q1['volume'].plot(legend=True,color='red')
plt.subplot(4,1,2)
plt.title('Amazon Stock Volume Analysis')
q2['volume'].plot(legend=True,color='b')
plt.subplot(4,1,3)
plt.title('Google Stock Volume Analysis')
q3['volume'].plot(legend=True,color='g')
plt.subplot(4,1,4)
plt.title('Microsoft Stock Volume Analysis')
q4['volume'].plot(legend=True,color='c')
plt.show()
```

```python
plt.rcParams['figure.figsize']=(50.0,15.0)
style.use('ggplot')
plt.subplots_adjust(hspace = .25)
plt.title ('Apple Closing Price Analysis')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.subplot(2,2,1)
plt.plot(q1.date,q1.close,color='red')
plt.plot(q2.date,q2.close,color='b')
plt.plot(q3.date,q3.close,color='g')
plt.plot(q4.date,q4.close,color='c')
plt.legend(df1.ticker.unique())
plt.show()
```

```python
#plotting percentage change for Apple's stock
df1['percentage_change]=df1['close'].pct_change()*100
df1
```

```python
dt=df1[df1.percentage_change > 1.0]
dt
```

```python
#percentage change for Apple's stock
pc = dt[dt['ticker']=='AAPL']
```

```python
plt.rcParams['figure.figsize']=(35.0,15.0)
style.use('ggplot')
plt.title('Apple\'s Percentage Change')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.plot(pc.date,pc.percentage_change,color='g')
plt.show()
```
